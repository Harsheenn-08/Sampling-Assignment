```python
import pandas as pd
import numpy as np

data = pd.read_csv("/content/Creditcard_data.csv")
data.head()
```

```python
data.shape
```

```python
data['Class'].isna().sum()
```

```python
X=data.drop('Class',axis=1)
y=data['Class']
```

```python
from collections import Counter

print("Original data:",Counter(y))
```

```python
#Sampling 1 using Random Under Sampling

from imblearn.under_sampling import RandomUnderSampler
Sample1 = RandomUnderSampler(random_state=42)
X_Sample1 , y_Sample1 = Sample1.fit_resample(X,y)

print("Sampling 1:",Counter(y_Sample1))
```

```python
#Sampling 2 using Random Over Sampling

from imblearn.over_sampling import RandomOverSampler
Sample2 = RandomOverSampler(random_state=42)
X_Sample2 , y_Sample2 = Sample2.fit_resample(X,y)

print("Sampling 2:",Counter(y_Sample2))
```

```python
#Sampling 3 using Tomek links: Under Sampling
from imblearn.under_sampling import TomekLinks
Sample3 = TomekLinks()
X_Sample3 , y_Sample3 = Sample3.fit_resample(X,y)

print("Sampling 3:",Counter(y_Sample3))
```

```python
from imblearn.over_sampling import SMOTE

Sample4 = SMOTE()

X_Sample4,y_Sample4 = Sample4.fit_resample(X,y)

print("Sampling 4:",Counter(y_Sample4))
```

```python
from imblearn.under_sampling import NearMiss

Sample5 = NearMiss()

X_Sample5,y_Sample5 = Sample5.fit_resample(X,y)
print("Sampling 5:",Counter(y_Sample5))
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler

X_train , X_test , y_train , y_test = train_test_split(
    X_Sample1 , y_Sample1, test_size = 0.2 , random_state=42 , stratify = y_Sample1)

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

M1 = LogisticRegression(C=1.0, penalty='l2',max_iter=1000)
M1.fit(X_train , y_train)

y_predict = M1.predict(X_test)
lraccuracy_Sample1 = accuracy_score(y_test, y_predict)
print("Logistic Regression for Sample 1:",lraccuracy_Sample1)


X_train , X_test , y_train , y_test = train_test_split(
    X_Sample2 , y_Sample2, test_size = 0.2 , random_state=42 , stratify = y_Sample2)

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

M1 = LogisticRegression(C=1.0, penalty='l2',max_iter=1000)
M1.fit(X_train , y_train)

y_predict = M1.predict(X_test)
lraccuracy_Sample2 = accuracy_score(y_test, y_predict)
print("Logistic Regression for Sample 2:",lraccuracy_Sample2)


X_train , X_test , y_train , y_test = train_test_split(
    X_Sample3 , y_Sample3, test_size = 0.2 , random_state=42 , stratify = y_Sample3)

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

M1 = LogisticRegression(C=1.0, penalty='l2',max_iter=1000)
M1.fit(X_train , y_train)

y_predict = M1.predict(X_test)
lraccuracy_Sample3 = accuracy_score(y_test, y_predict)
print("Logistic Regression for Sample 3:",lraccuracy_Sample3)


X_train , X_test , y_train , y_test = train_test_split(
    X_Sample4 , y_Sample4, test_size = 0.2 , random_state=42 , stratify = y_Sample4)

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

M1 = LogisticRegression(C=1.0, penalty='l2',max_iter=1000)
M1.fit(X_train , y_train)

y_predict = M1.predict(X_test)
lraccuracy_Sample4 = accuracy_score(y_test, y_predict)
print("Logistic Regression for Sample 4:",lraccuracy_Sample4)


X_train , X_test , y_train , y_test = train_test_split(
    X_Sample5 , y_Sample5, test_size = 0.2 , random_state=42 , stratify = y_Sample5)

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

M1 = LogisticRegression(C=1.0, penalty='l2',max_iter=1000)
M1.fit(X_train , y_train)

y_predict = M1.predict(X_test)
lraccuracy_Sample5 = accuracy_score(y_test, y_predict)
print("Logistic Regression for Sample 5:",lraccuracy_Sample5)
```
